Lemmatization

Lemmatization is a text normalization technique in Natural Language Processing (NLP) that reduces words to their dictionary base form (lemma) using linguistic knowledge such as vocabulary and grammar rules. Unlike stemming, which simply chops off suffixes, lemmatization ensures the output is a valid word.

```python
#sample words to apply lemmatization

words = ['eating', 'eats', 'eaten', 'writing', 'writes', 'programming', 'programs', 'history', 'goes']
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
#wordnet_lemmatizer object

wordnet_lemmatizer = WordNetLemmatizer()

for word in words:
    print(f"{word}------->{wordnet_lemmatizer.lemmatize(word)}")

```

WordNetLemmatizer().lemmatize() has a argument pos (parts of speech) based on which different lemma is given as output

* pos="n"--noun
* pos="v"--verb
* pos="a"--adjective
* pos="r"--adverb

```python
#example with change in pos tag

lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("going", pos="n")
```

```python
lemmatizer.lemmatize("going", pos="v")
```
