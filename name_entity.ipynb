### Name Entity Recognition

```python
import nltk

sentence = """The Effile Tower was built from 1887 to 1889 by French Engineer Gustave Effile, whose company specialized in building metal frameworks and structures.
Person eg: Mounica
Place or Location: India
Date eg: December 12
Time: 12:05 PM
Money: 1 million dollar
Organization: Alpha Private Limited
Percent: 20% twenty percent"""
```

```python
list_of_words=nltk.word_tokenize(sentence)
```

```python
pos_tags = nltk.pos_tag(list_of_words)
pos_tags
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.ne_chunk(pos_tags).draw()
```
