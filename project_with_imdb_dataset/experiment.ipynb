### IMDB Movie Review Classification using Simple RNN

```python
#importing libraries
import tensorflow
from tensorflow.keras.datasets import imdb
```

```python
max_features=10000
train_data, test_data= imdb.load_data(num_words=max_features, start_char=1, index_from=3, oov_char=2)
train_reviews, train_labels = train_data
test_reviews, test_labels = test_data
print(f"shape of train_reviews {train_reviews.shape} and its labels of shape {train_labels.shape}")
print(f"shape of test_reviews {test_reviews.shape} and its labels of shape {test_labels.shape}")
```

```python
#inspecting a sample review and it's label

print(train_reviews[0])
print(train_labels[0])
```

```python
word_index = imdb.get_word_index()
reverse_word_index = {value+3:key for key, value in word_index.items()}
```
