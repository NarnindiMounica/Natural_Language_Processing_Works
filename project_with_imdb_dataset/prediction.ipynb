### Prediction pipeline

```python
#importing libraries

import numpy as np
import pandas as pd

from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import load_model
```

```python
word_index = imdb.get_word_index()
reverse_word_index = {value: key for (key, value) in word_index.items()}
```

```python
#loading the model

model = load_model('model.keras')
model.summary()
```

```python
#sample review for prediction
import re
sample_review = "This movie was fantastic! The acting was great and the plot was thrilling"

#custom helper functions
def preprocess_review(sample_review:str):
    sample_review = sample_review.lower()
    sample_review = re.sub('[^a-zA-Z]', ' ', sample_review)
    sample_review_words = sample_review.split(" ")
    sample_review_seq = [word_index.get(word , -1 ) + 3 for word in sample_review_words]
    padded_sample_review_seq = sequence.pad_sequences([sample_review_seq], maxlen=500)
    return padded_sample_review_seq

def decode_review(review_seq:list)->str:
    review_sent = " ".join([reverse_word_index.get(index-3, "?") for index in review_seq])
    return review_sent
```
