### Embedding Layer Implementation

```python
#importing libraries

import tensorflow
from tensorflow.keras.preprocessing.text import one_hot #this is deprecated 
from tensorflow.keras.layers import TextVectorization

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
#sample sentences

sentences = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'i am a good girl',
        'i am a good developer',
        'understand the meaning of words',
        'your videos are good']
```

```python
voc_size=10000
one_hot_representation = [one_hot(sent, voc_size) for sent in sentences]
one_hot_representation
```

```python
from tensorflow.keras.layers import TextVectorization

'''tf.keras.layers.TextVectorization(
    max_tokens=None,
    standardize='lower_and_strip_punctuation',
    split='whitespace',
    ngrams=None,
    output_mode='int',
    output_sequence_length=None,
    pad_to_max_tokens=False,
    vocabulary=None,
    idf_weights=None,
    sparse=False,
    ragged=False,
    encoding='utf-8',
    name=None)'''

vectorize_layer = TextVectorization(max_tokens=voc_size, output_mode="int")
vectorize_layer.adapt(sentences)
```

```python
sent_vectors=vectorize_layer(sentences)
```

```python
#ragged vectors

vectorize_layer = TextVectorization(max_tokens=voc_size, ragged=True)

vectorize_layer.adapt(sentences)

input_sentences = vectorize_layer(sentences)
input_sentences
```

```python
# passing ragged vector through pad_sequences

sequences = [sen.numpy() for sen in input_sentences]
padded_sequences=pad_sequences(sequences, maxlen=5)
padded_sequences
```

```python
#word embedding representation

model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=10, input_length=5))

model.summary()
```

```python
model.compile(loss="mse", optimizer="adam")
```

```python
model(sent_vectors)
```
