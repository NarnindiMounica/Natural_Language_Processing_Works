### Tokenization

```python
#importing libraries

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, TreebankWordTokenizer
```

```python
corpus="My name is Mounica and I'm learning NLP. I am interested in learing AI skills and aspire to become AI Engineer"

sentences = sent_tokenize(text=corpus)
sentences
```

* It splits corpus into list of sentences

```python
type(sentences)
```

```python
#word tokenization

for sentence in sentences:
    print(word_tokenize(sentence))
```

* it splits each sentences into list of words.

```python
#corpus with punctuations included

corpus = "She bought apples, oranges, bananas, and grapes; however, she forgot the milk. Wow! Did you really finish the project on time?"
```

```python
#applying treebank tokenizer

treebank_tokenizer = TreebankWordTokenizer()

treebank_tokenizer.tokenize(corpus)
```

This tokenizer performs the following steps:

* split standard contractions, e.g. don't -> do n't and they'll -> they 'll
* treat most punctuation characters as separate tokens
* split off commas and single quotes, when followed by whitespace
* separate periods that appear at the end of line
