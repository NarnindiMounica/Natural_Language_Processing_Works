### Bag of Words Implementation Using Spam Classification Dataset

```python
import pandas as pd
import nltk

from sklearn.feature_extraction.text import CountVectorizer

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
dataset = pd.read_csv("datasets\SMSSpamCollection",delimiter="\t",names=['classification','message'])
```

```python
dataset
```

```python
#text preprocessing on message feature

#lowering case
dataset['message'] = dataset['message'].str.lower()

#replacing any char apart from letters with blank space
dataset['message'] = dataset['message'].str.replace('[^a-zA-Z]', " ", regex=True)
```

```python
dataset['message'].head(5)
```

```python
#removing stopwords and applying lemmatization on it

wordnet_lemmatizer = WordNetLemmatizer()
corpus = []
for i in range(len(dataset['message'])):
    review = dataset['message'][i].split(" ")
    review = [wordnet_lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review = " ".join(review)
    corpus.append(review)

```
