### Spam Classifier Using BOW (Bag of Words) & TF-IDF (Term Frequency - Inverse Document Frequency)

```python
#importing libraries

import nltk
import pandas as pd
import numpy as np

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
dataset = pd.read_csv("datasets\SMSSpamCollection", delimiter="\t", names=['label', 'message' ])
```

```python
dataset.head(2)
```

```python
#data preprocessing
corpus = []
for item in range(len(dataset['message'])):
    review = dataset['message'][item].lower()
    review = dataset['message'][item].replace("[^a-zA-Z]", " ").split(" ")
    review = [PorterStemmer().stem(word) for word in review if word not in set(stopwords.words("english"))]
    review = " ".join(review)
    corpus.append(review)
```

```python
#defining target

y = pd.get_dummies(dataset['label'], dtype=np.int)
y
```
