### Spam Classifier Using BOW (Bag of Words) & TF-IDF (Term Frequency - Inverse Document Frequency)

##### Using Bag of Words:

```python
#importing libraries

import nltk
import pandas as pd
import numpy as np

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

from sklearn.model_selection import train_test_split

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, classification_report
```

```python
dataset = pd.read_csv("datasets\SMSSpamCollection", delimiter="\t", names=['label', 'message' ])
```

```python
dataset.head(2)
```

```python
#data preprocessing
corpus = []
for item in range(len(dataset['message'])):
    review = dataset['message'][item].lower()
    review = dataset['message'][item].replace("[^a-zA-Z]", " ").split(" ")
    review = [PorterStemmer().stem(word) for word in review if word not in set(stopwords.words("english"))]
    review = " ".join(review)
    corpus.append(review)
```

```python
#defining target

y = pd.get_dummies(dataset['label'], dtype=np.int32)
y = y.iloc[:, 0]
y.head(5)
```

```python
#splitting into train and test sets

x_train, x_test, y_train, y_test = train_test_split(corpus, y, test_size=0.2, random_state=12)
```

```python
#converting words to vector using BOW

count_vectorizer = CountVectorizer(max_features=100, ngram_range=(1,1), binary=True)

#applying on train and test input text

x_train_arr = count_vectorizer.fit_transform(x_train).toarray()
x_test_arr = count_vectorizer.transform(x_test).toarray()
```

```python
#training model using features

multinomial_nb = MultinomialNB()
multinomial_nb.fit(x_train_arr, y_train)
y_pred = multinomial_nb.predict(x_test_arr)
```

```python
#performance metrics evaluation

accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
print(classification_report(y_test, y_pred))
```
