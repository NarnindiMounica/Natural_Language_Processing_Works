### Stemming and It's Types

Stemming is a text preprocessing technique in Natural Language Processing (NLP) that reduces words to their root form by stripping suffixes and prefixes, helping simplify text and improve efficiency in tasks like search, classification, and information retrieval. 

#### Porter Stemmer

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
words = ['eating', 'eats', 'eaten', 'writing', 'writes', 'programming', 'programs', 'history', 'goes']

#applying porterstemmer on above list of words

#stemmer object
porter_stemmer = PorterStemmer()

for word in words:
    print(f"{word}------------------->{porter_stemmer.stem(word)}")
```

#### Regexp Stemmer

The RegexpStemmer is a simple, rule-based stemming approach that uses regular expressions to strip suffixes or prefixes from words

```python
from nltk.stem import RegexpStemmer
```

```python
#applying regexp stemmer on given list of words

#regexp stemmer object

#removes common suffixes like ing|es|s from words.
regexp_stemmer = RegexpStemmer(regexp='ing$|es$|s$')

for word in words:
    print(f"{word}--------------->{regexp_stemmer.stem(word)}")
```
