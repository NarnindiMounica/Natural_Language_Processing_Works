### Word2Vec & Avg Word2Vec Implementation

```python
#importing libraries

import gensim, re
from gensim.models import Word2Vec, KeyedVectors

import gensim.downloader as api

from nltk.stem import WordNetLemmatizer
import pandas as pd
import numpy as np

from tqdm import tqdm
```

```python
wv = api.load("word2vec-google-news-300")

vec_king = wv['king']
```

```python
#vocabulary list
print(f"Number of words in vocabulary: {len(wv.index_to_key)}")
wv.index_to_key
```

```python
#finetuning it with smsspam classifier

dataset = pd.read_csv("datasets\\SMSSpamCollection", delimiter="\t", names=['label', 'message'])
dataset.head()
```

```python
wordnet_lemmatizer = WordNetLemmatizer()

corpus = []
for item in range(len(dataset['message'])):

    review = dataset['message'][item].lower()
    review = re.sub('[^a-zA-Z]', " ", review).split(" ")
    review = [wordnet_lemmatizer.lemmatize(word) for word in review ]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#another way to preprocess messages.

from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess

#simple_preprocess : split sentence into words, converts to lower case and ignore too short or too long tokens

words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words
```

```python
#training word2vec from scratch

model = gensim.models.Word2Vec(words)
model.wv.index_to_key
```

```python
def avg_word_to_vec(doc):

    return(np.mean([model.wv[word] for word in doc  if word in model.wv.index_to_key], axis=0))
    
```

```python
#applying custom function on every sentence

x = []
for i in tqdm(range(len(words))):
    x.append(avg_word_to_vec(words[i]))
```
